### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
# My code
import requests

requestGet = requests.get("https://www.lifesupplier.com.tw")
requestGet.encoding = "UTF-8"
print(requestGet.text)
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
import requests

header = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.79 Safari/537.36'}
requestGet = requests.get("https://www.lifesupplier.com.tw", headers = header)
requestGet.encoding = "UTF-8"
print(requestGet.text)
```
