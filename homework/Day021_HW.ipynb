## 1. 取出今天所有的新聞

```python
# My Code
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# 每 0.01 秒鐘自動往下滑
import time
for i in range(1000):
    time.sleep(0.01)
    browser.execute_script("window.scrollTo(0, 1000000);")
```

```python
# 取得資料，丟到 BeautifulSoup
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    if '07/09' in d.find(class_="date").text:
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
count = 1

for d in soup.find(class_="part_list_2").find_all('h3'):
    if count == 1:
        timeStart = d.find(class_="date").text
        timeStart = timeStart[11:len(timeStart)]
        timeStart = str(int(timeStart[:2])-2) + timeStart[2:len(timeStart)]
    else :
        timeData = d.find(class_="date").text
        timeData = timeData[11:len(timeData)]
        
    if count > 1:
        if timeStart in d.find(class_="date").text or (int(timeData[:2]) < int(timeStart[:2]) and int(timeData[3:5]) < int(timeStart[3:5])):
            break
            
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
    count += 1
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# My Code
from selenium.webdriver.support.ui import Select

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

# Year
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2020')

# Month
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('7')

# Data
selectData = Select(browser.find_element_by_id("selD"))
selectData.select_by_value('7')

# Click Button
browser.find_element_by_id('button').click()
```

```python
# 每 0.01 秒鐘自動往下滑
for i in range(450):
    time.sleep(0.01)
    browser.execute_script("window.scrollTo(0, 1000000);")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
# My Code
timeStart = '15:00'
timeEnd   = '17:00'

for d in soup.find(class_="part_list_2").find_all('h3'):

    timeData = d.find(class_="date").text
    timeData = timeData[11:len(timeData)]
        
    
    if int(timeData[:2]) < int(timeStart[:2]):
        break
    if int(timeData[:2]) < int(timeEnd[:2]) or timeData == '17:00':    
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
