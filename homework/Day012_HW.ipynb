### ① 取出今天所有的發文

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

# print(r.text[:66000])
df = pd.DataFrame(columns=['時間', '分類', '標題'])
i=0
for d in soup.find(class_="part_list_2").find_all('h3'):
    df.loc[i] = [d.find(class_="date").text, d.find_all('em')[-1].text,d.find_all('a')[-1].text]
    print(d.find(class_="date").text, d.find_all('em')[-1].text,d.find_all('a')[-1].text)
    i+=1
    
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df.sort_values('分類')
```

### ③ 哪一個類別的文章最多

```python
df[['分類','標題']].groupby('分類').count().sort_values(['標題'], ascending=False).head(10)
```
